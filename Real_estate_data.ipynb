```python
result_container = soup.find_all('li',{'class':'SearchResultsList__WideCell-b7y9ki-2'})
```

```python
results_update=[]  
```

```python
 for r in result_container:
    if r.has_attr('data-testid'):
        results_update.append(r)
```

```python
len(results_update)
```

```python
results_update
```

```python
url_part_1='https://www.trulia.com/CA/San_Diego'
```

```python
url_part_2=[]

for item in results_update:
    
    for link in item.find_all('div', {'data-testid':'property-card-details'}):
        url_part_2.append(link.find('a').get('href'))
```

```python
url_part_2
```

```python
url_joined=[]
for link_2 in url_part_2:
    url_joined.append(urllib.parse.urljoin(url_part_1,link_2))
```

```python
first_link = url_joined[1]
```

```python
response5 = requests.get(first_link)
```

```python
soup=BeautifulSoup(response5.content,'html.parser')
```

```python
address = []
bedrooms = []
bathrooms =[]
area = []
year_built = []
parking = []
price = []

for link in url_joined:
    resp = requests.get(link)
    
    soup1 = BeautifulSoup(resp.content,'html.parser')
    
    try :
        address.append(soup1.find('span',{'data-testid':'home-details-summary-headline'}).get_text())
    except:
        address.append('')
    try :
        bedrooms.append(soup1.find('li',{'data-testid':'bed'}).get_text())
    except:
        bedrooms.append('')    
    try :
        bathrooms.append(soup1.find('li',{'data-testid':'bath'}).get_text())
    except:
        bathrooms.append('')
    try :
        area.append(soup1.find('li',{'data-testid':'floor'}).get_text())
    except:
        area.append('')   
    try :
        year_built.append(soup1.find('div',string='Year Built').findNext('div').get_text())
    except:
        year_built.append('')
    try :
        parking.append(soup1.find('div', string='Parking').findNext('div').get_text())
    except:
        parking.append('')  
    try :
        price.append(soup1.find('h3',{'data-testid':"on-market-price-details"}).get_text())
    except:
        price.append('')    
    
    output = {'Address':address,'Bedrooms':bedrooms,'Bathrooms':bathrooms,'Area':area,'Year_Built':year_built,'Parking':parking,'Price':price}    
        
```
